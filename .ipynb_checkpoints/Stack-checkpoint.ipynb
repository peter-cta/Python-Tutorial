## Stack : Ngăn xếp
- Tuân thủ nguyên tắc : Last in - First out
- Ứng dụng :
    + Stack được sử dụng nhiều trong trình biên dịch để biên dịch các ngôn ngữ lập trình
    + Stack còn được dùng để kiểm tra tính chính xác của công thức toán học, công thức hóa học , ví dụ việc đóng mở ngoặc tròn
        - 4 * (5 + 3)
        - (COO)2H3

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self,elem):
        self.stack.append(elem)
    def pop(self):
        if len(self.stack) > 0:            
            self.stack.pop()
        else:
            print('Stack is Empty!')
    def printAllElement(self):
        if len(self.stack) > 0:
            for i in self.stack:
                print(i,end='')
        else:
            print('Stack is Empty')
    
        
```

## Leetcode 20 : Valid Parentheses
Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.
An input string is valid if:
    - Open brackets must be closed by the same type of brackets.
    - Open brackets must be closed in the correct order.

Example 1: 
    Input: s = "()"
    Output: true

Example 2:
    Input: s = "()[]{}"
    Output: true

Example 3:
    Input: s = "(]"
    Output: false

Example 4:
    Input: s = "([)]"
    Output: false

Example 5:
    Input: s = "{[]}"
    Output: true

```python
class Solution:
    def isValid(self, s: str) -> bool:
        dau = {"{":"}", "[":"]", "(":")"}
        stack = []
        for i in s:
            if i in dau:
                stack.append(i)
            else:
                if len(stack) == 0:
                    return False
                elif dau[stack[-1]] == i:
                    stack.pop()
                else :
                    return False
        if len(stack) == 0:
            return True
```
