## Tuple
- Tuple cũng giống như List gồm một chuỗi các phần tử nhưng bất biến 

### 1. Khởi tạo một Tuple()

```python
tuple1 = ('Chu The Anh', '05-09-1999', 'Hung Yen', 21, 'PTIT')
tuple1
```

```python
tuple2 = 'Chu The Anh', 21, 'Hung Yen', 'PTIT'
tuple2
```

```python
tuple3 = tuple()
tuple3
```

### 2. Truy cập các phần tử trong tuple()

```python
tuple4 = ('Chu The Anh', 21, 'Hung Yen', 'PTIT')
tuple4[0]
```

```python
tuple4[1]
```

```python
tuple4[:2]
```

### 3. Tuple là bất biến

```python
tuple5 = ('Quan', 'Ao', 'Giay')
tuple6 = ('Sach', 'Vo', 'But')

tuple5 + tuple6
```

### 4. Các hàm cơ bản trong tuple

- Count() : Trả về số lượng phần tử xuất hiện trong tuple

```python
tuple7 = (1, 3, 4, 5, 6 ,7, 1, 3, 5)
tuple7.count(1)
```

- Index() : Trả về chỉ số của phần tử xuất hiện đầu tiên

```python
tuple8 = (1, 3, 4, 5, 7, 3)
tuple8.index(3)
```

- map() : Trả về một list kết quả dựa vào điều kiện cho trước

```python
tuple9 = (1, 2, 3, 4, 5)
def square(x):
    return x**2
list(map(square, tuple9))
```

- Filter() : Trả về một List gồm kết quả lọc theo điều kiện cho trước

```python
tuple10 = ('a', 'b', 'c', '1', '2', '3')
def filterAlphabet(x):
    return x.isalpha()

list(filter(filterAlphabet, tuple10))
```

```python
tuple11 = ('a', 'b', 'c', '1', '2', '3')
def filterDigit(x):
    return x.isdigit()

list(filter(filterDigit, tuple11))
```

```python
- for in, in, while , enumerate, reversed : có chức năng tương tự như đối với List
- Zip() : Nhóm các phần tử trong tuple có cùng vị trí lại với nhau , tao ra một list các tuple mới
```

```python
tuple12 = ('tay', 'chân', 'mũi')
tuple13 = ('2', '2', '1')

list(zip(tuple12, tuple13))
```

- Sum() : sẽ trả về tổng của một tuple gồm các số , trả ra lỗi khi có kí tự xuất hiện


```python
tuple14 = (22, 33, 44)
sum(tuple14)
```

```python
tuple15 = (22, 33, 44, 'a')
sum(tuple15)
```
