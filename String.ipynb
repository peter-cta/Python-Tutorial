### String : Chuỗi
   - Là một kiểu dữ liệu trong python cũng là dang bất biến và không thể thay đổi

#### Một số hàm cơ bản trong xử lý chuỗi

```python
 # Capitalize() : Viết hoa chữ cái đầu trong chuỗi
str1 = 'chu the anh'
str1.capitalize()
```

```python
# Join() : Trả về một chuỗi gồm các phần tử của list đc nối với điều kiện cho trước ( chỉ thực hiện nối kí tự , không nối số)
list1 = ['a', 'b', 'c', 'd']
str2 = ','
str2.join(list1)
```

```python
# % : Ghép biên vào chuỗi 
name = "Chu The Anh"
str3 = 'Ten toi la : %s' %(name)
str3
```

##### UNICODE

```python
u_string = u'Tôi yêu Việt Nam'
u_string
```

```python
# convert unicode thành dạng byte 'utf-8'
b = u_string.encode('utf-8')
b
```

```python
# Convert từ 'utf-8' sang dạng unicode
u = b.decode('utf-8')
u
```
